```python
import requests
from bs4 import BeautifulSoup
from IPython.core.display import display, HTML
```

```python
#url = "https://factfinder.census.gov/faces/tableservices/jsf/pages/productview.xhtml?src=bkmk"
url = "http://www.gopsusports.com/sports/m-basebl/stats/2017-2018/teamcume.html"
result = requests.get(url)
print(result.status_code)
```

```python
html = result.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
p = soup.find(name='p')
table = p.find('table')
print(table.text)
```

```python
keys = []
for row in table.findAll(name='tr')[0:1]:
    for key in row.findAll(name='td')[1:]:
        keys.append(key.text.encode('ascii','ignore'))
print(keys)
```

```python
dictionary = {}
for rows in table.findAll(name='tr')[1:]:
    #print(rows)
    #print('**************')
    if rows.name == 'tr':
        try:
            i = 0
            player = str(rows.find('a').text.encode('ascii'))
            dictionary[player] = {}
            for column in rows.findAll('td')[1:]:
                key = str(keys[i])
                dictionary[player][key] = str(column.text.encode('ascii','ignore'))
                i = i+1
        except: 
            pass
        
```

```python
print(dictionary)
```
